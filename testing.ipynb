```python
from model import evolve_params, gen_pdf, compute_dist, gen_domain
import matplotlib.pyplot as plt
```

```python
lower=1
mode=3
upper=8
resolution=1000
a_mut, loc_mut, scale_mut = 2, 0.05, 0.05
n_mutations=50
edge_target = 0.025
iters=50
edge_exp = 4
modal_exp = 2
```

```python
params = {"a": 1, "loc": mode, "scale": ((mode - lower) + (upper - mode)) / 2}
X = gen_domain(lower, upper, params, resolution)
```

```python
selected, logs = evolve_params(
    lower=lower,
    mode=mode,
    upper=upper,
    resolution=resolution,
    a_mut = a_mut,
    loc_mut = loc_mut,
    scale_mut = scale_mut,
    n_mutations=n_mutations,
    iters=iters,
    edge_target=edge_target,
    edge_exp = edge_exp,
    modal_exp = modal_exp
)
```
